```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def euler(f, t0, y0, T, h):

    # f = funcao da EDO
    # t0 = valor inicial do tempo
    # y0 = valor inicial da solucao
    # T = tempo final da simulacao
    # h = tamanho do passo
    
    # calcula o numero total de passos com base no intervalo de tempo e tamanho do passo
    N = int((T - t0) / h)
    
    # cria um vetor de tempos de t0 ate T com N+1 pontos igualmente espaçados
    t = np.linspace(t0, T, N+1)
    
    # inicializa o vetor y com zeros 
    y = np.zeros(N+1)
    
    # define o valor inial
    y[0] = y0
    
    for n in range(N):

        # atualiza y usando a formula do metodo de Euler:
        y[n+1] = y[n] + h * f(t[n], y[n])
    
    # retorna os vetores de tempo e solucao aproximada
    return t, y
```

```python
# PVI 1

# derivada dada no PVI
f1 = lambda t, y: y - t**2 + 1 

# solucao exata
y_exact1 = lambda t: (t + 1)**2 - 0.5 * np.exp(t) 

# parametros 
t0, y0, T, h = 0, 0.5, 2, 0.2 

# aplicacao do metodo
t1, y_num1 = euler(f1, t0, y0, T, h)

# solucao exata
y_true1 = y_exact1(t1)

# erro absoluto
erro_abs1 = np.abs(y_true1 - y_num1)

# erro relativo
erro_rel1 = erro_abs1 / np.abs(y_true1)

# tabela de erros
df1 = pd.DataFrame({
    't': t1,
    'y_exato': y_true1,
    'y_euler': y_num1,
    'erro_abs': erro_abs1,
    'erro_rel': erro_rel1
})
df1
```

```python
# grafico do PVI 1
plt.figure(figsize=(10, 5))
plt.plot(t1, y_true1, label='Solucao exata', linewidth=2)
plt.plot(t1, y_num1, 'o--', label='Euler', linewidth=2)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('PVI 1: Solucao exata vs Euler')
plt.legend()
plt.grid()
plt.show()

# grafico dos erros
plt.figure(figsize=(10, 4))
plt.plot(t1, erro_abs1, 'r-o', label='Erro absoluto')
plt.plot(t1, erro_rel1, 'g-s', label='Erro relativo')
plt.xlabel('t')
plt.title('Erros no PVI 1')
plt.legend()
plt.grid()
plt.show()
```

```python
# PVI 2

# derivada dada no PVI
f2 = lambda t, y: y + t 

# solucao exata
y_exact2 = lambda t: 2 * np.exp(t) - t - 1 

# aplicacao do metodo
t2, y_num2 = euler(f2, 0, 1, 1.5, 0.1)

# solucao exata
y_true2 = y_exact2(t2)

# erro absoluto
erro_abs2 = np.abs(y_true2 - y_num2)

# erro relativo
erro_rel2 = erro_abs2 / np.abs(y_true2)

# tabela de erros
df2 = pd.DataFrame({
    't': t2,
    'y_exato': y_true2,
    'y_euler': y_num2,
    'erro_abs': erro_abs2,
    'erro_rel': erro_rel2
})
df2
```

```python
# grafico do PVI 2
plt.figure(figsize=(10, 5))
plt.plot(t2, y_true2, label='Solucao exata', linewidth=2)
plt.plot(t2, y_num2, 'o--', label='Euler', linewidth=2)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('PVI 2: Solucao exata vs Euler')
plt.legend()
plt.grid()
plt.show()

# grafico dos erros
plt.figure(figsize=(10, 4))
plt.plot(t2, erro_abs2, 'r-o', label='Erro absoluto')
plt.plot(t2, erro_rel2, 'g-s', label='Erro relativo')
plt.xlabel('t')
plt.title('Erros no PVI 2')
plt.legend()
plt.grid()
plt.show()
```
