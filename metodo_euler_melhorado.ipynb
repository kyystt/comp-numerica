```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def euler_melhorado(f, t0, y0, T, h):

    # calcula o numero de passos com base no intervalo total e no tamanho do passo
    N = int((T - t0) / h)

    # cria um vetor de tempos t com N+1 pontos igualmente espa√ßados entre t0 e T
    t = np.linspace(t0, T, N+1)

    # inicializa o vetor se solucoes com zeros
    y = np.zeros(N+1)

    # define o valor inicial
    y[0] = y0
    
    for n in range(N):
        # calcula a inclinacao inicial com base no ponto atual
        k1 = f(t[n], y[n])

        # estima a inclinacao futura com base no valor estimado de y no proximo passo
        k2 = f(t[n] + h, y[n] + h * k1)

        # atualiza y[n+1] usando a media ponderada de k1 e k2
        y[n+1] = y[n] + h * (k1 + k2) / 2
    
    # retorna os vetores de tempo e solucao aproximada
    return t, y
```

```python
# PVI 1

# derivada do PVI 1 
f1 = lambda t, y: y - t**2 + 1 

# solucao exata
y1_exact = lambda t: (t+1)**2 - 0.5*np.exp(t) 

# parametros
t0, y0, T, h = 0, 0.5, 2, 0.2

# aplicacao do metodo
t, y_numerico = euler_melhorado(f1, t0, y0, T, h)

# solucao exata
y_exato = y1_exact(t)

# erro absoluto
erro_abs1 = np.abs(y_exato - y_numerico)

# erro relativo
erro_rel1 = erro_abs1 / np.abs(y_exato)

# tabela dos erros
pd.DataFrame({
    't': t,
    'y_exato': y_exato,
    'y_RK2': y_numerico,
    'Erro Absoluto': erro_abs1,
    'Erro Relativo (%)': 100 * erro_rel1
})
```

```python
# grafico do PVI 1
plt.figure(figsize=(10,5))
plt.plot(t, y_exato, 'b-', label='Solucao exata')
plt.plot(t, y_numerico, 'ro--', label='RK2')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Exemplo 1: y\' = y - t^2 + 1')
plt.legend()
plt.grid(True)
plt.show()

# graficos dos erros
plt.figure(figsize=(10, 4))
plt.plot(t, erro_abs1, 'r-o', label='Erro absoluto')
plt.plot(t, erro_rel1, 'g-s', label='Erro relativo')
plt.xlabel('t')
plt.ylabel('Erro')
plt.title('Erros no PVI 1')
plt.legend()
plt.grid()
plt.show()
```

```python
# PVI 2

# derivada de PVI 2
f2 = lambda t, y: -2*t*y 

# solucao exata
y2_exact = lambda t: np.exp(-t**2) 

# parametros
t0, y0, T, h = 0, 1, 2, 0.2

# aplicacao do metodo
t, y_numerico2 = euler_melhorado(f2, t0, y0, T, h)

# solucao exata
y_exato2 = y2_exact(t)

# erro absoluto
erro_abs2 = np.abs(y_exato2 - y_numerico2)

# erro relativo
erro_rel2 = erro_abs2 / np.abs(y_exato2)

# tabela dos erros
pd.DataFrame({
    't': t,
    'y_exato': y_exato2,
    'y_RK2': y_numerico2,
    'Erro Absoluto': erro_abs2,
    'Erro Relativo (%)': 100 * erro_rel2
})
```

```python
# grafico do PVI 2
plt.figure(figsize=(10,5))
plt.plot(t, y_exato2, 'b-', label='Solu√ß√£o exata')
plt.plot(t, y_numerico2, 'ro--', label='RK2')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Exemplo 2: y\' = -2ty')
plt.legend()
plt.grid(True)
plt.show()

# grafico dos erros 
plt.figure(figsize=(10, 4))
plt.plot(t, erro_abs2, 'r-o', label='Erro absoluto')
plt.plot(t, erro_rel2, 'g-s', label='Erro relativo')
plt.xlabel('t')
plt.ylabel('Erro')
plt.title('Erros no PVI 1')
plt.legend()
plt.grid()
plt.show()
```
